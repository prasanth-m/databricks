```
data = "dbfs:/FileStore/tables/in_vehicle_coupon_recommendation.csv"
df = spark.read.format("csv").option("header","true")\
  .option("inferSchema", "true").load(data)
```

```
train_df, test_df = df.randomSplit([0.80, 0.20], seed=3)
```

```
display(test_df)
```

```
from databricks import automl
summary = automl.classify(train_df, target_col="Y", timeout_minutes=15)
```

```
help(summary)
```

```
model_uri = summary.best_trial.model_path
print(model_uri)
```

```
import mlflow
 
# Prepare test dataset
test_pdf = test_df.toPandas()
y_test = test_pdf["Y"]
X_test = test_pdf.drop(["Y"], axis=1)
X_test.head()
```

```
# Run inference using the best model
model = mlflow.pyfunc.load_model(model_uri)
predictions = model.predict(X_test)
test_pdf["coupon_accepted"] = predictions
```

```
test = X_test.fillna(0)
```

```
# Run inference using the best model
model = mlflow.pyfunc.load_model(model_uri)
predictions = model.predict(test)
test_pdf["coupon_accepted"] = predictions
```

```
import sklearn.metrics
 
model = mlflow.sklearn.load_model(model_uri)
sklearn.metrics.plot_confusion_matrix(model, X_test, y_test)
```
